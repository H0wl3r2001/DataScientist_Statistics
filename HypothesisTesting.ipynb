# Hypothesis Testing

Hypothesis testing is a fundamental concept in statistics used to make inferences about populations based on sample data.
In this notebook, we will explore different hypothesis tests, their assumptions, and how they are applied in real-world scenarios.

## 1. Null vs Alternative Hypothesis
- **Null Hypothesis (H₀):** Assumes no effect or no difference.
- **Alternative Hypothesis (H₁):** Represents what we seek to prove.

For example, in a drug effectiveness study:
- **H₀:** The drug has no effect.
- **H₁:** The drug has a positive effect.

### 1.1 Rejection Region and Significance level

- The **significance level** ($\alpha$) is the probability of rejecting H₀ when it is actually true.
- Common choices are 0.05 (5%) and 0.01 (1%).
- The **rejection region** is where we reject H₀ if the test statistic falls into it.

```python
# Visualizing Rejection Region
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)
plt.plot(x, y)
plt.fill_between(x, y, where=(x > 1.96), color='red', alpha=0.5, label='Rejection Region')
plt.fill_between(x, y, where=(x < -1.96), color='red', alpha=0.5)
plt.title("Rejection Region for α=0.05 in a Two-Tailed Test")
plt.legend()
plt.show()
```
