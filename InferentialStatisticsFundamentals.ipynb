# Inferential Statistics Fundamentals

Inferential Statistics involves the use of a sample to estimate some characteristic in a large population; and to test a research hypothesis about a given population. This notebook explores fundamental initial concepts of inferential statistics. It covers distributions, the normal distribution, the standard normal distribution, the Central Limit Theorem, standard error, and estimation techniques.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

## 1. Distributions (recap)

As discussed in the [DataScientist_Probability](https://github.com/H0wl3r2001/DataScientist_Probability) project, a probability distribution describes how values of a random variable are distributed. It can be:
- **Discrete** (e.g., Binomial, Poisson)
- **Continuous** (e.g., Normal, Exponential)

A distribution provides insights into the likelihood of different outcomes:

```python
# Demonstrating a Uniform Distribution
sample = np.random.uniform(0, 1, 1000)
sns.histplot(sample, bins=20, kde=True)
plt.title("Uniform Distribution")
plt.show()
```

### 1.1 The Normal Distribution

The normal distribution is a symmetric, bell-shaped distribution characterized by its mean ($\mu$) and standard deviation ($\sigma$).

The probability density function (PDF) is given by:
$$
f(x) = \frac{1}{\sigma\sqrt{2\pi}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}
$$

```python
# Generating a Normal Distribution
normal_sample = np.random.normal(loc=50, scale=10, size=1000)
sns.histplot(normal_sample, bins=30, kde=True)
plt.title("Normal Distribution")
plt.show()
```

#### 1.1.1 The Standard Normal Distribution

A normal distribution with mean 0 and standard deviation 1 is called the standard normal distribution (Z-distribution).

To standardize a value $x$:
$$
z = \frac{x - \mu}{\sigma}
$$

```python
# Standardizing a normal sample
z_scores = (normal_sample - np.mean(normal_sample)) / np.std(normal_sample)
sns.histplot(z_scores, bins=30, kde=True)
plt.title("Standard Normal Distribution")
plt.show()
```

## 2. Sampling Theory and Estimation

Estimation theory is a branch of statistics and signal processing that deals with estimating the values of parameters based on measured/empirical data that has a random component. This branch encompasses concepts such as Central Limit Theorem (how sample statistics behave), Standard Error (the variability of sample means) and Estimators and Estimates (methods to infer population parameters from samples).

The following sections will cover these exact terms and concepts.

### 2.1 Central Limit Theorem

The Central Limit Theorem (CLT) states that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the population's distribution.

```python
# Demonstrating CLT
means = [np.mean(np.random.uniform(0, 1, 30)) for _ in range(1000)]
sns.histplot(means, bins=30, kde=True)
plt.title("Central Limit Theorem in Action")
plt.show()
```

### 2.2 Standard Error

The standard error (SE) quantifies the variability of sample means:

$$
SE = \frac{\sigma}{\sqrt{n}}
$$

It decreases as sample size increases, improving estimate accuracy.

```python
sample_sizes = np.arange(1, 101)
standard_errors = 10 / np.sqrt(sample_sizes)
plt.plot(sample_sizes, standard_errors)
plt.xlabel("Sample Size")
plt.ylabel("Standard Error")
plt.title("Standard Error vs Sample Size")
plt.show()
```

### 2.3 Estimators and Estimates

The definition of these terms are, respectively:

- A rule or formula used to estimate a parameter.
- The specific value obtained from applying an estimator to a sample.

Some common estimators include:

- **Sample mean ($\bar{x}$)** estimates population mean ($\mu$).
- **Sample variance ($s^2$)** estimates population variance ($\sigma^2$).

```python
sample_data = np.random.normal(50, 10, 100)
estimated_mean = np.mean(sample_data)
estimated_variance = np.var(sample_data, ddof=1)
print(f"Estimated Mean: {estimated_mean}")
print(f"Estimated Variance: {estimated_variance}")
```
